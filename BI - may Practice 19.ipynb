```python
salesf = pd.ExcelFile('Sales.Sample.xlsx')
```

```python
salesfile = pd.read_excel(salesf,'sales')
orderRfile = pd.read_excel(salesf,'returned')
productRfile = pd.read_excel(salesf,'productrating')
```

```python
join1 = salesfile.join(orderRfile,how='left',lsuffix='ORDERNUMBER',rsuffix='ORDERNUMBER')
```

```python
join1= join1.drop(columns= ['ORDERNUMBERORDERNUMBER'])
```

```python
join1['RETURNED']=join1['RETURNED'].fillna('Not Returned')
```

```python
join2 = join1.join(productRfile,how='left',lsuffix='PRODUCTCODE',rsuffix='PRODUCTCODE')
```
