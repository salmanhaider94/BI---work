```python
# renaming cols to make it more meaning full
df= df.rename(columns={'GRADES_PK_G':'GRADES_Pre-Kindergarten', 'GRADES_KG_G':'GRADES_KINDERGARTEN',
       'GRADES_4_G':'GRADES_4', 'GRADES_8_G':'GRADES_8', 'GRADES_12_G':'GRADES_12', 'GRADES_1_8_G':'GRADES_1 till 8',
       'GRADES_9_12_G':'GRADES_9 till 12', 'GRADES_ALL_G':'GRADES_ALL'})
```

```python
df.shape
```

```python
df.head()
```

```python
# dropping because it's just and combination of state and year which is already given - it won't be needed in analysis
df.drop(columns=['PRIMARY_KEY'])
```

```python
#filling missing values with mean
df['TOTAL_REVENUE']=df['TOTAL_REVENUE'].fillna(df['TOTAL_REVENUE'].mean())
df['TOTAL_EXPENDITURE']=df['TOTAL_EXPENDITURE'].fillna(df['TOTAL_EXPENDITURE'].mean())
df['AVG_MATH_4_SCORE']=df['AVG_MATH_4_SCORE'].fillna(df['AVG_MATH_4_SCORE'].mean())
df['AVG_MATH_8_SCORE']=df['AVG_MATH_8_SCORE'].fillna(df['AVG_MATH_8_SCORE'].mean())
df['AVG_READING_4_SCORE']=df['AVG_READING_4_SCORE'].fillna(df['AVG_READING_4_SCORE'].mean())
df['AVG_READING_8_SCORE']=df['AVG_READING_8_SCORE'].fillna(df['AVG_READING_8_SCORE'].mean())



```

```python
df.head()
```

```python
df= df.rename(columns={'AVG_MATH_4_SCORE':'GRADE_4_MATH_SCORE','AVG_MATH_8_SCORE':'GRADE_8_MATH_SCORE','AVG_READING_4_SCORE':'GRADE_4_READING_SCORE','AVG_READING_8_SCORE':'GRADE_8_READING_SCORE'})
```
